##1. Escribir una función que tome un texto como entrada y devuelva la cantidad de palabras únicas en ese texto, ignorando mayúsculas, minúsculas y signos de puntuación.


```python
import string

def contar_palabras_unicas(texto):
    # Convertir a minúsculas
    texto = texto.lower()
    # Eliminar signos de puntuación
    texto = texto.translate(str.maketrans('', '', string.punctuation))
    # Dividir en palabras
    palabras = texto.split()
    # Contar palabras únicas
    palabras_unicas = set(palabras)
    return len(palabras_unicas)


# Ejemplo de uso:
for i, texto in enumerate(corpus):
  num_palabras_unicas = contar_palabras_unicas(texto)
  print(f"Texto {i+1}: Número de palabras únicas: {num_palabras_unicas}")
```

##2. Desarrollar una función que tome un texto y un valor entero 'n' como entrada, y genere todos los n-gramas (secuencias de 'n' palabras contiguas) presentes en el texto.

```python
def generar_ngramas(texto, n):
    palabras = texto.split()
    ngramas = []
    for i in range(len(palabras) - n + 1):
        ngramas.append(palabras[i:i+n])
    return ngramas

for i, texto in enumerate(corpus):
  print(f"Texto {i+1}: {generar_ngramas(texto, 3)}")
```

##3. Crear una función que tome un texto y dos palabras como entrada, y reemplace todas las ocurrencias de la primera palabra con la segunda en el texto.

##4. Implementar una función que tome un texto y una lista de palabras como entrada, y devuelva un diccionario que mapee cada palabra de la lista a la cantidad de veces que aparece en el texto.

```python
def contar_ocurrencias_palabras(texto, lista_palabras):
    conteo = {}
    for palabra in lista_palabras:
        conteo[palabra] = texto.lower().split().count(palabra.lower())
    return conteo


# Ejemplo de uso:
for i, texto in enumerate(corpus):
  print(f"Texto {i+1}: {contar_ocurrencias_palabras(texto, ['ciencia', 'datos', 'es'])}")
```

##5. Escribir una función que tome un texto como entrada y devuelva una lista de palabras ordenadas por su frecuencia de aparición en el texto, de mayor a menor.

```python
import string
from collections import Counter

def ordenar_palabras_por_frecuencia(texto):
    texto = texto.lower()
    texto = texto.translate(str.maketrans('', '', string.punctuation))
    palabras = texto.split()
    conteo_palabras = Counter(palabras)
    palabras_ordenadas = sorted(conteo_palabras, key=conteo_palabras.get, reverse=True)
    return palabras_ordenadas


# Ejemplo de uso:
for i, texto in enumerate(corpus):
  print(f"Texto {i+1}: {ordenar_palabras_por_frecuencia(texto)}")
```

##6. Desarrollar una función que tome un texto y una lista de palabras como entrada, y devuelva el texto sin las palabras que estén en la lista.

```python
def eliminar_palabras(texto, lista_palabras):
    palabras_texto = texto.split()
    texto_filtrado = [palabra for palabra in palabras_texto if palabra.lower() not in [p.lower() for p in lista_palabras]]
    return " ".join(texto_filtrado)


# Ejemplo de uso:
for i, texto in enumerate(corpus):
  print(f"Texto {i+1}: {eliminar_palabras(texto, ['la', 'sobre', 'de', 'el'])}")
```

##7. Implementar una función que tome un texto como entrada y devuelva un diccionario que mapee la longitud de las palabras a la cantidad de palabras de esa longitud en el texto.

```python
import string

def contar_palabras_por_longitud(texto):
    # Convertir a minúsculas y eliminar signos de puntuación
    texto = texto.lower()
    texto = texto.translate(str.maketrans('', '', string.punctuation))
    # Dividir en palabras
    palabras = texto.split()
    # Contar palabras por longitud
    conteo_por_longitud = {}
    for palabra in palabras:
        longitud = len(palabra)
        if longitud in conteo_por_longitud:
            conteo_por_longitud[longitud] += 1
        else:
            conteo_por_longitud[longitud] = 1
    return conteo_por_longitud


# Ejemplo de uso:

for i, texto in enumerate(corpus):
  print(f"Texto {i+1}: {contar_palabras_por_longitud(texto)}")
```

##8. Escribir una función que tome un texto como entrada y devuelva una versión del texto donde las palabras repetidas consecutivas se reduzcan a una única ocurrencia.

```python
def eliminar_palabras_repetidas(texto):
    palabras = texto.split()
    texto_sin_repetidas = []
    if palabras:
        texto_sin_repetidas.append(palabras[0])
        for i in range(1, len(palabras)):
            if palabras[i] != palabras[i-1]:
                texto_sin_repetidas.append(palabras[i])
    return " ".join(texto_sin_repetidas)

for i, texto in enumerate(corpus):
  print(f"Texto {i+1}: {eliminar_palabras_repetidas(texto)}")
```

##9. Desarrollar funciones para codificar y decodificar un texto utilizando técnicas simples como la sustitución de caracteres o el cifrado César.

El **cifrado César** es una técnica de cifrado clásica que consiste en desplazar cada letra de un texto determinado un número fijo de posiciones en el alfabeto. Por ejemplo, si usamos un desplazamiento de 3, la letra 'A' se convertirá en 'D', la 'B' en 'E', la 'C' en 'F' y así sucesivamente. Este desplazamiento se realiza de manera circular, es decir, si llegamos al final del alfabeto, volvemos al principio.

Ejemplo:

Texto =  "hola mundo"

Desplazamiento: 3 posiciones}

Texto cifrado = "krod pxqgr"

```python
def codificar_texto(texto, desplazamiento):
    resultado = ""
    for caracter in texto:
        if 'a' <= caracter <= 'z':
            codigo_ascii = ord('a') + (ord(caracter) - ord('a') + desplazamiento) % 26
            resultado += chr(codigo_ascii)
        elif 'A' <= caracter <= 'Z':
            codigo_ascii = ord('A') + (ord(caracter) - ord('A') + desplazamiento) % 26
            resultado += chr(codigo_ascii)
        else:
            resultado += caracter
    return resultado

def decodificar_texto(texto_codificado, desplazamiento):
    resultado = ""
    for caracter in texto_codificado:
        if 'a' <= caracter <= 'z':
            codigo_ascii = ord('a') + (ord(caracter) - ord('a') - desplazamiento) % 26
            resultado += chr(codigo_ascii)
        elif 'A' <= caracter <= 'Z':
            codigo_ascii = ord('A') + (ord(caracter) - ord('A') - desplazamiento) % 26
            resultado += chr(codigo_ascii)
        else:
            resultado += caracter
    return resultado


# Ejemplo de uso:
texto = corpus[9]  # Utilizamos el décimo texto del corpus
texto_codificado = codificar_texto(texto, 3)
print("Texto codificado:", texto_codificado)
texto_decodificado = decodificar_texto(texto_codificado, 3)
print("Texto decodificado:", texto_decodificado)
```

##10. Crear una función que tome un texto como entrada y devuelva la cantidad de frases presentes en ese texto, asumiendo que las frases están separadas por puntos, signos de interrogación o exclamación.

```python
import re

def contar_frases(texto):
    # Usamos una expresión regular para encontrar los delimitadores de frase
    frases = re.split(r'[.?!]+', texto)
    # Eliminamos los posibles elementos vacíos al final de la lista
    frases = [frase for frase in frases if frase.strip()]
    return len(frases)


# Ejemplo de uso:
texto = corpus[6]  # Utilizamos el séptimo texto del corpus
print(texto)
print(contar_frases(texto))
```
